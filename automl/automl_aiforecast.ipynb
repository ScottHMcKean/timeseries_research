```python
ts_table = 'shm.timeseries.m5_processed'
```

```python
from databricks import automl
summary = automl.forecast(
  dataset=spark.table(ts_table), 
  target_col="y", 
  time_col="ds",
  identity_col="unique_id", 
  frequency="d",
  horizon=3, 
  timeout_minutes=30,
  )
```
